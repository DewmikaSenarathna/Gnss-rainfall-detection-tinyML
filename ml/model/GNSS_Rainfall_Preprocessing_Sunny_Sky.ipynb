```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/DATA.CSV")
```

```python
print(df.info())
print(df.head())
```

```python
zero_ratio = (df == 0).mean() * 100
print(zero_ratio.sort_values(ascending=False))
```

```python
all_zero_cols = [col for col in df.columns if (df[col] == 0).all()]
df = df.drop(columns=all_zero_cols)

print("Dropped columns:", all_zero_cols)
```

```python
snr_cols = [col for col in df.columns if col.startswith("SNR")]
```

```python
df[snr_cols] = df[snr_cols].replace(0, np.nan)
```

```python
df['Temp(C)'] = df['Temp(C)'].replace(0, np.nan)
df['Humidity(%)'] = df['Humidity(%)'].replace(0, np.nan)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")

numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = imputer.fit_transform(df[numeric_cols])
```

```python
df["RainLabel"] = 0
```

```python
print(df.columns.tolist())
```

```python
df.describe().T
```

```python
df.to_csv("Sunny_day_clean_dataset.csv", index=False)
```
