```python
import pandas as pd
import os
```

```python
sunny_path = "/content/Sunny_Clean.csv"
low_rain_path = "/content/Low_Rain_Clean.csv"
high_rain_path = "/content/Rainy_Havvy_Clean.csv"
```

```python
sunny_df = pd.read_csv(sunny_path)
low_rain_df = pd.read_csv(low_rain_path)
high_rain_df = pd.read_csv(high_rain_path)
```

```python
sunny_df["WeatherLabel"] = 0
low_rain_df["WeatherLabel"] = 1
high_rain_df["WeatherLabel"] = 2
```

```python
print("Sunny columns:", sunny_df.columns)
print("LowRain columns:", low_rain_df.columns)
print("HighRain columns:", high_rain_df.columns)
```

```python
merged_df = pd.concat(
    [sunny_df, low_rain_df, high_rain_df],
    axis=0,
    ignore_index=True
)
```

```python
merged_df = merged_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
print("Shape:", merged_df.shape)
print("\nClass distribution:")
print(merged_df["WeatherLabel"].value_counts())

print("\nMissing values:")
print(merged_df.isnull().sum())
```

```python
output_path = "GNSS_Merged_Dataset.csv"
merged_df.to_csv(output_path, index=False)

print("Merged dataset saved successfully.")
```
